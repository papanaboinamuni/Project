**Treat the Missing values using imputation method**

```python
# FIll the Card_Category value with mode 
mode_value = Data["Card_Category"].mode()[0]

# Fill missing values with the mode value
Data["Card_Category"].fillna(value=mode_value, inplace=True)
```

```python
#Fill the Credit_Limit value with mean value
mean_value = Data["Credit_Limit"].mean()

# Fill missing values with the mode value
Data["Credit_Limit"].fillna(value=mean_value, inplace=True)
```

```python
#Check the null values 
Data.isnull().sum()
```

```python
#Summarize the data
Data.describe()
```

```python
# Outliers in the dataset
Q1 = Data.quantile(0.25)
Q3 = Data.quantile(0.75)
IQR = Q3 - Q1
outliers = ((Data < (Q1 - 1.5 * IQR)) | (Data > (Q3 + 1.5 * IQR))).sum()
print("Outliers present in the dataset:")
print(outliers)
```
