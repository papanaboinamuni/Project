```python
# Sord the data in descending order based on revenue and filter the top  500 movies
sorted_Data=Drop_rows.sort_values('revenue',ascending=False)
sorted_Data.head(500)
```

```python
#Define a function for central tendency 
def info_of_numerical(col):
    print(f"The mean of the {col} is {sorted_Data[col].mean()}")
    print(f"The median of the {col} is {sorted_Data[col].median()}")
    print(f"The mode of the {col} is {sorted_Data[col].mode()[0]}")
    print(f"The standard deviation of the {col} is {sorted_Data[col].std()}")
    print(f"Number of missing values in the {col} is {sorted_Data[col].isnull().sum()}")
```

```python
# find the measure of central Tendency for budget column
info_of_numerical("budget")
```

```python
# Visualize the boxplot for outlier analysis
plt.figure(figsize=(10,5))
plt.boxplot(sorted_Data['budget'],vert=False,patch_artist=True,boxprops=dict(facecolor='green'))
plt.title('Budget Box Plot')
plt.show()
```

**Interpretation:**

    The central tendency of the budget column shows the average of 40654444, median value of 25000000 and mode value shows the 20000000. And visualise purpose box plot is used, in that I can observe that outliers are present in the upperfence and the mean value is greater median value so itâ€™s a right skewness.
